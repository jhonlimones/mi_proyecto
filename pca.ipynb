```python
# Dividir datos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(featurs, labels, test_size=0.33, random_state=42)
```

```python
label_encoder = LabelEncoder()

# Convertir las etiquetas categÃ³ricas ('e' y 'p') a numÃ©ricas (0 y 1)
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Predecir el pipeline

def train_predict(learner, sample_size, X_train, y_train, X_test, y_test):

    results = {}
    start = time()
    learner = learner.fit(X_train[:sample_size], y_train[:sample_size])
    end = time()

    results['train_time'] = end - start

    start = time()
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train[:300])
    end = time()

    results['pred_time'] = end - start
    results['acc_train'] = accuracy_score(y_train[:300],predictions_train)
    results['acc_test'] = accuracy_score(y_test,predictions_test)
    results['f_train'] = fbeta_score(y_train[:300], predictions_train, beta=0.5)
    results['f_test'] = fbeta_score(y_test,predictions_test, beta=0.5)
    print('{} trained on - {} samples'.format(learner.__class__.__name__, sample_size))
    return results
```

ELECCION DEL MEJOR MODELO

Goussain Neive Bayes
Random Forest
KNN

```python
# Librerias importadas

Gauss_A = GaussianNB()
Random_B = RandomForestClassifier()
knn_C = KNeighborsClassifier()

training_length = len(X_train)
samples_1 = int(training_length * 0.01)
samples_10 = int(training_length * 0.1)
samples_100 = int(training_length * 1)

# Diccionario para almacenar datos
results = []

# Entrenamiento de los modelos y almacenamiento de los resultados
for model in [Gauss_A, Random_B, knn_C]:
    model_name = model.__class__.__name__
    for i, samples in enumerate([samples_1, samples_10, samples_100]):
        # Capturamos la precisiÃ³n devuelta por la funciÃ³n `train_predict`
        metrics = train_predict(model, samples, X_train, y_train_encoded, X_test, y_test_encoded)
        # Obtenemos solo la precisiÃ³n del conjunto de prueba
        accuracy = metrics['acc_test']  # Usamos 'acc_test' para la precisiÃ³n en el test set
        # Agregamos los resultados al diccionario
        results.append({
            'Model': model_name, # Nombre del modelo
            'Fold': i + 1,       # Para indicar el fold 1, 2, o 3      
            'Samples': samples,  # numero de muestras fold
            'Accuracy': accuracy * 100 # Asumiendo que accuracy es un entre 0 y 1 multiplicamos por 100 para porcentaje
        })
# Crear un DataFrame a partir de los resultados
df_results = pd.DataFrame(results)

# Mostrar el DataFrame completo
print(df_results)

# Encontrar el modelo con mejor desempeÃ±o (maxima precision)
best_model = df_results.loc[df_results['Accuracy'].idxmax()]

# Mostrar el mejor modelo
print('\nMejor Modelo:')
print(best_model)
```
